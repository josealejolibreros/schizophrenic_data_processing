```python
# Example
#todos los tres grupos juntos
cor = data_encoded.corr()
idx_strs = []
r2_values = []


for i in range(1,24):
    for j in range(25,189):
        # Independent variables
        
        #Funcion
        x = data_encoded.columns[i]
        y = 'edad'

        # Dependent variables
        
        #Estructura
        z = data_encoded.columns[j]

        # Pairings
        xz = cor.loc[ x, z ]
        yz = cor.loc[ y, z ]
        xy = cor.loc[ x, y ]


        Rxyz = math.sqrt((abs(xz**2) + abs(yz**2) - 2*xz*yz*xy) / (1-abs(xy**2)) )
        R2 = Rxyz**2

        # Calculate adjusted R-squared
        n = len(data_encoded) # Number of rows
        k = 2       # Number of independent variables
        R2_adj = 1 - ( ((1-R2)*(n-1)) / (n-k-1) )
        idxs_str = str(x + ',' + y + ',' + z)
        idx_strs.append(idxs_str)
        r2_values.append(R2_adj)
        #print('Coef( ' + idxs_str + ') = ' + str(R2_adj))
   
#max_r2 = np.max(np.array(r2_values))
#[print("Mayor coef: ", str(idx_strs[i]), ": ", str(r2_values[i])) for i, j in enumerate(r2_values) if j == max_r2]

print("Greater than 0.4: ")
for i, j in enumerate(r2_values):
    if j >=0.4:
        tipo_grupo_todos_registros.append("sanos_esquizofrenicos_dementes")
        variables_correlacion_todos_registros.append(idx_strs[i])
        valores_correlacion_todos_registros.append(r2_values[i])
        print(str(idx_strs[i]), ": ", str(r2_values[i])) 
```
