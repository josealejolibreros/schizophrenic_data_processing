```python
def define_title(combination_vars, side, diagnostic, type_matrix):
    if combination_vars == matching_amygdala_thalamus:
        title = type_matrix + ": " + "Amygdala - Thalamus\n"
    elif combination_vars == matching_amygdala_hippocampus:
        title = type_matrix + ": " + "Amygdala - Hippocampus\n"
    elif combination_vars == matching_hippocampus_thalamus:
        title = type_matrix + ": " + "Hippocampus - Thalamus\n"
    
    if side == 1:
        title = title + "Left side\n"
    elif side == 2:
        title = title + "Right side\n"
        
    if diagnostic == 0:
        title = title + "Control group"
    elif diagnostic == 1:
        title = title + "Schizophrenia no dementia"
    elif diagnostic == 2:
        title = title + "Schizophrenia with dementia"
        
    return title
```

```python
def detect_repeated(str1, str2):
    '''
    Specific order:
    
    Amygdala - talamo
    Amygdala - hippocampus
    Hippocampus - talamo
    
    Anything else is a repeated tuple
    '''
    
    if ("talamo" in str1 and "talamo" in str2) or \
    ("hippocampus" in str1 and "hippocampus" in str2) or \
    ("amygdala" in str1 and "amygdala" in str2) or \
    ("talamo" in str1 and "amygdala" in str2) or \
    ("hippocampus" in str1 and "amygdala" in str2) or \
    ("talamo" in str1 and "hippocampus" in str2):
        repeated = True
    else:
        repeated = False
        
    return repeated
        
```

```python
def extract_corr_submatrix(current_combination_vars, corr_matrix):
    corr_matrix = np.array(corr_matrix)
    if current_combination_vars == matching_amygdala_thalamus:
        corr_matrix = corr_matrix[0:len(matching_amygdala)-1, len(matching_amygdala):]
    elif current_combination_vars == matching_amygdala_hippocampus:
        corr_matrix = corr_matrix[0:len(matching_amygdala)-1, len(matching_amygdala):]
    elif current_combination_vars == matching_hippocampus_thalamus:
        corr_matrix = corr_matrix[0:len(matching_hippocampus)-1, len(matching_hippocampus):]
    
    return corr_matrix

def get_label_names(current_combination_vars):
    if current_combination_vars == matching_amygdala_thalamus:
        ynames = matching_amygdala
        xnames = matching_thalamus
    elif current_combination_vars == matching_amygdala_hippocampus:
        ynames = matching_amygdala
        xnames = matching_hippocampus
    elif current_combination_vars == matching_hippocampus_thalamus:
        ynames = matching_hippocampus
        xnames = matching_thalamus
    
    return xnames, ynames
```
